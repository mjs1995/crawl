```python
import time
from selenium import webdriver
from itertools import count
from bs4 import BeautifulSoup
import pandas as pd 
import urllib.request as REQ
```

```python
url = "http://www.menupan.com/restaurant/bestrest/bestrest.asp?page={}&trec=925&areacode=jj201&pt=rt"

results = []

for n in range(1,38):
    link = url.format(n)
    response = REQ.urlopen(link)
    bs = BeautifulSoup(response, 'html.parser')
    tag_body = bs.select_one('.rankingList .list')
    lis = tag_body.findAll('li')
    
    for li in lis:
        rank = li.select_one('.numTop, .rankNum').text
        listName = li.select_one('.listName').string
        listType = li.select_one('.listType').string
        listArea = li.select_one('.listArea').string
        results.append({    "랭킹" : rank,
                            "상호명" : listName,
                            "업종" : listType,
                            "지역" : listArea})
        df = pd.DataFrame(results)
```
